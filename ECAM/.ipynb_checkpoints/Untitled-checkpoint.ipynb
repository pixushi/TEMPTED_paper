```python
import os
import csv
import pandas as pd
import numpy as np
from mprod import table2tensor
from mprod.dimensionality_reduction import TCAM


# reading in the txt data table
npc = int(3)
ss = str(int(1))
jj = str(int(9))
# obtain meta data and otu table
file_path = 'simdata/sim'+ss+'_ntime'+jj+'.csv'
data_table = pd.read_csv(file_path, usecols=lambda col: col != 'delivery', \
              dtype={'month':int, 'studyid':str})
data_table = data_table.iloc[:, 2:]
data_table = data_table.sample(n=100)
data_table = data_table.set_index(['studyid', 'month'])
data_table
```

```python
# format into data tensor
data_tensor, map1, map3 =  table2tensor(data_table)
```

```python


# run TCAM
tca = TCAM()
tca_trans = tca.fit_transform(data_tensor)

# save results
tca_df = pd.DataFrame(tca_trans)
tca_df.rename(index = dict(map(reversed, map1.items())), inplace = True)   
tca_df.iloc[:,:3]
```
